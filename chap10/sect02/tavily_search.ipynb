```python
from langchain_tavily import TavilySearch

tavily_search = TavilySearch(max_result=5)
```

```python
res = tavily_search.invoke({'query': "2026 한국 경제 전망"})

# print(res['results'])

for r in res['results']:
    print(r)
```

```python
from IPython.display import JSON

JSON(res['results']).data
```

```python
from tavily import TavilyClient

client = TavilyClient()

content = client.search(
    "2026년 한국 경제 전망",
    search_depth="advanced",
    include_raw_content=True
)

JSON(content).data
```

```python
from langchain_openai import ChatOpenAI

llm = ChatOpenAI(model="gpt-4o")

# 질문에 대한 자료를 TavilyClient로 가져옴 
query = '현대자동차 미국 시장 2026년 전망'

content = client.search(
    query, 
    include_raw_content=True,
    search_depth='advanced'
)["results"]
```

```python
prompt =[{
    "role": "system",
    "content": (
        f'정보: """{content}"""\n\n'
        f'위의 정보를 사용하여, 다음 질문에 대해 자세한 보고서를 한국어로 작성하세요.: "{query}"\n'
        '-신문 기사 형식을 사용하되, MLA 표준을 준수하는 markdown 문법을 사용해주세요.'
        '-활용한 자료는 출처를 명시하세요.'
    )
}]
```

```python
report = llm.invoke(prompt).content
print(report)
```
